```python
import pandas as pd
import random

apparel_products = [
    "Graphic T-Shirt", "Pullover Hoodie", "Hooded Sweatshirt",
    "Comfort Fit Joggers", "Drawstring Joggers", "Jogger Sweatpants",
    "Tank Top", "Crewneck Sweatshirt", "Denim Jacket", "Cardigan"
]

accessories_products = [
    "Baseball Cap", "Leather Wallet", "Sunglasses", "Belt",
    "Beanie", "Scarf", "Wristwatch", "Backpack", "Handbag", "Keychain"
]

categories = {
    "Apparel": apparel_products,
    "Accessories": accessories_products
}

products = []

for i in range(1000):
    category = random.choice(list(categories.keys()))
    product_name = f"Classic {random.choice(['Red', 'Blue', 'Black', 'Gray', 'Navy', 'White', 'Green'])} {random.choice(categories[category])}"
    price = random.randint(10, 250)  
    products.append({
        "title": product_name,
        "price": price,
        "category": category
    })

df = pd.DataFrame(products)

df.to_csv("../datasets/mongo-data.csv", index=False)

print("1,000 dummy products generated successfully!")
```
